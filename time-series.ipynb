**1. The relationship between the penguin's bill length and bill depth for each species.** (5 marks)

```python
## TODO: Check your attributes, make a graph using Matplotlib or Seaborn to show the relationship needed

plt.figure(figsize=(12, 8))
sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species")

plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.title("Relationship between Penguin's Bill Length and Bill Depth for Each Species")

plt.legend(title="Species")
plt.grid(True)
plt.tight_layout()
plt.show()
```

 **What do you observe from the graph?** (2 marks)

Your answer here:


From the scatter plot, I observed that:

1. Each species forms distinct clusters or groups in the plot, indicating that penguins of the same species tend to have similar bill lengths and bill depths.

2. The clusters for different species are separated from each other, suggesting that there are differences in bills between species. For example, Adelie penguins generally have shorter and deeper bills compared to Gentoo and Chinstrap penguins.

3. There appears to be a positive correlation between bill length and bill depth for all species, as seen by the general trend of points sloping upwards from left to right. This suggests that penguins with longer bills also tend to have deeper bills, and vice versa.

**2. Suitable graph of the mean BMI (BMI not body_mass weight_in_kg/height(m2)) of female and male penguins. (Assume the height is the flipper length)**  (5 marks)


```python
#Your answer here

penguins_with_bmi = penguins.copy()
#BMI for each penguin

penguins_with_bmi['bmi'] = (penguins_with_bmi['body_mass_g'] / 1000) / ((penguins_with_bmi['flipper_length_mm'] / 1000) ** 2)

#Calculate mean BMI for female and male penguins
mean_bmi_by_sex = penguins_with_bmi.groupby('sex')['bmi'].mean().reset_index()
print(mean_bmi_by_sex)

# Create a bar plot
plt.figure(figsize=(5, 6))
sns.barplot(data=mean_bmi_by_sex, x='sex', y='bmi', palette='Set2')

# Add labels and title
plt.xlabel("Sex")
plt.ylabel("Mean BMI")
plt.title("Mean BMI of Female and Male Penguins")

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

 **What do you observe from the graph?** (2 marks)

Your answer here:


1. There is only a slight difference in the mean BMI between female and male penguins. 

2. Typically, male penguins on average have a higher BMI than female penguins.


**Why did you choose this type of graph?** (2 marks)

I chose a bar plot for visualizing the mean BMI of female and male penguins because:

1. Bar plots are effective for comparing data across different groups or categories. In this case, we're comparing the mean BMI between female and male penguins, which are two distinct groups.

2. Bar plots provide a clear and straightforward representation of the mean values for each group. The height of the bars directly corresponds to the mean BMI of each group.

3. The sex of the penguins (female and male) is a categorical variable, and bar plots are well-suited for visualizing distributions or summaries of categorical data.
