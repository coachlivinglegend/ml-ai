```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import accuracy_score

from sklearn.model_selection import cross_val_score, KFold
```

# Part A: Model Training and Evaluation (25 Points)

## Loading the data

```python
# Load the data
x_train = pd.read_csv('./data/x_train.csv')
y_train = pd.read_csv('./data/y_train.csv').squeeze()
x_test = pd.read_csv('./data/x_test.csv')
y_test = pd.read_csv('./data/y_test.csv').squeeze()
```

##### Choose a shallow (simple) classification baseline model from the ones mentioned in class after the midterm. A shallow classifier is one that doesn’t generate a complex decision boundary. We will use this classifier in the following points. Include a comment on why you choose this particular classifier and why is it considered shallow.

I'll use a simple decision Tree—specifically, a decision tree with limited depth. Decision trees are considered shallow when they have constraints like maximum depth or minimum samples per leaf, which prevents them from creating very complex decision boundaries. This makes them fast to train and easy to understand.

```python
# Create a decision tree classifier as the base model
base_clf = DecisionTreeClassifier(max_depth=3)  # Limit depth to make the tree shallow
```

##### Train the chosen baseline model on the training set. Make sure to use 5-fold cross-validation to properly judge the performance of the model on a validation set.

```python
from sklearn.model_selection import cross_val_score, KFold

# Define the cross-validation strategy
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Evaluate the baseline model using cross-validation
baseline_scores = cross_val_score(base_clf, x_train, y_train, cv=cv, scoring='accuracy')
print(f"Baseline Decision Tree CV Accuracy: {baseline_scores.mean():.4f} ± {baseline_scores.std():.4f}")
```

##### Utilize the bagging method, with the base estimator being the model chosen in the first step, to do a training run on the training set. Make sure to use 5-fold cross-validation to properly judge the performance of the model on a validation set.

```python
# Setup the Bagging Classifier
bagging_clf = BaggingClassifier(estimator=base_clf, n_estimators=20, random_state=42)

# Evaluate the bagging ensemble using cross-validation
bagging_scores = cross_val_score(bagging_clf, x_train, y_train, cv=cv, scoring='accuracy')
print(f"Bagging Ensemble CV Accuracy: {bagging_scores.mean():.4f} ± {bagging_scores.std():.4f}")
```

##### Utilize the boosting method, with the base estimator being the model chosen in the first step, to do a training run on the training set. Make sure to use 5-fold cross-validation to properly judge the performance of the model on a validation set.

```python
# Setup the Boosting Classifier
boosting_clf = AdaBoostClassifier(estimator=base_clf, n_estimators=20, random_state=42)

# Evaluate the boosting ensemble using cross-validation
boosting_scores = cross_val_score(boosting_clf, x_train, y_train, cv=cv, scoring='accuracy')
print(f"Boosting Ensemble CV Accuracy: {boosting_scores.mean():.4f} ± {boosting_scores.std():.4f}")
```

##### Validate the trained baseline model, trained bagging and boosting models on the testing set and compare their performance. Include a comment on which method worked better and a possible reasoning behind the observations.

```python
# Train and evaluate the baseline model
base_clf.fit(x_train, y_train)
baseline_test_accuracy = accuracy_score(y_test, base_clf.predict(x_test))

# Train and evaluate the bagging ensemble
bagging_clf.fit(x_train, y_train)
bagging_test_accuracy = accuracy_score(y_test, bagging_clf.predict(x_test))

# Train and evaluate the boosting ensemble
boosting_clf.fit(x_train, y_train)
boosting_test_accuracy = accuracy_score(y_test, boosting_clf.predict(x_test))

print(f"Baseline Test Accuracy: {baseline_test_accuracy:.4f}")
print(f"Bagging Test Accuracy: {bagging_test_accuracy:.4f}")
print(f"Boosting Test Accuracy: {boosting_test_accuracy:.4f}")
```

##### Comment

Bagging appears to be the best method for this particular dataset, providing a modest improvement over the baseline. Bagging is effective here likely because it reduces the variance of the model without significantly increasing the bias, which is useful given the stability of the baseline model’s performance.

Boosting does not perform as well due to its sensitivity to noise and outliers in the data. Boosting's approach of focusing intensely on misclassified instances can lead to overfitting, particularly with a dataset that may include some noise or non-representative data points.

# Part B: Guessing the Mystery Label (25 Points)

##### Train a simple neural network (MLP) with 3 hidden layers having the following number of neurons [256,128,64] on the training set. Make sure to normalize the data beforehand with z-score normalization.

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier

# Normalize the data
scaler = StandardScaler()
x_train_norm = scaler.fit_transform(x_train)
x_test_norm = scaler.transform(x_test)

# Define the MLP
mlp = MLPClassifier(hidden_layer_sizes=(256, 128, 64), max_iter=200, random_state=42)

# Train the MLP
mlp.fit(x_train_norm, y_train)
```

##### Utilize the uploaded helper code in “helper.py” to generate an embeddings matrix representing the output of the last hidden layer in the MLP for the test dataset.

```python
from helper import get_hidden_layer_activations

embeddings = get_hidden_layer_activations(mlp, x_test_norm, 2)
embeddings
```

##### Carry out PCA and LDA to reduce the dimensionality of the generated embeddings to 2. Produce a plot of the data in two dimensions for PCA and LDA, using easily distinguishable colours and markers to indicate the labels of each datapoint. Include a comment on any interesting patterns you see in the two plots.

```python
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import matplotlib.pyplot as plt

# PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(embeddings)

# LDA
lda = LDA(n_components=2)
lda_result = lda.fit_transform(embeddings, y_test)

# Visualization
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=y_test, cmap='viridis', marker='o', label='PCA ')
plt.title('PCA of Embeddings')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(lda_result[:, 0], lda_result[:, 1], c=y_test, cmap='viridis', marker='o', label='LDA')
plt.title('LDA of Embeddings')
plt.legend()
plt.show()
```

##### Carry out K-Means on the generated embeddings with 5 clusters and visualize the results by using the resulting clusters as alternate colour mappings for the PCA plot above.

```python
from sklearn.cluster import KMeans

# K-Means
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
clusters = kmeans.fit_predict(embeddings)


# Visualization
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=clusters, cmap='viridis', marker='o', label='PCA with Clusters')
plt.title('PCA of Embeddings')
plt.legend()
```

##### Based on the results seen in the plots, can you guess what are the labels for the given dataset (what each label number represents in terms of the category of clothes)? It might also help to list out a random selection of data entries (the original images) for each cluster and their label value to help understand the patterns each cluster might represent.

```python
def plot_images_for_cluster(cluster_id):
    indices = np.where(clusters == cluster_id)[0]
    # Select a few random indices to plot
    selected_indices = np.random.choice(indices, 10, replace=False)
    if isinstance(x_test, pd.DataFrame):

        images = x_test.iloc[selected_indices].to_numpy().reshape(-1, 28, 28)
    else:

        images = x_test[selected_indices].reshape(-1, 28, 28)

    plt.figure(figsize=(10, 5))
    for i, image in enumerate(images):
        plt.subplot(2, 5, i + 1)
        plt.imshow(image, cmap='gray')
        plt.axis('off')
    plt.show()
```
